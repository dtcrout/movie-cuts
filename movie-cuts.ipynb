```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['figure.figsize'] = (20, 10)

# Import data from csv
df = pd.read_csv('cinemetrics.csv', encoding='utf-8')

# Drop all null 
df = df.dropna()

df = df[['title',
         'year',
         'director',
         'country',
         'asl',
         'msl',
         'std']]

df.head(10)
```

```python

countries = df[['country']]

countries.head(10)
```

```python
def boxplot(df, country):
    tempDF = df[df['country'] == country]
    tempDF = tempDF[tempDF.year.apply(lambda x: all(y.isdigit() for y in x))]
    sns.boxplot(x='year', y='asl', data=tempDF, showfliers=False)
    plt.show()
```

```python
boxplot('USA')
```

```python
boxplot('France')
```
